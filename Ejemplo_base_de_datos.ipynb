#Para poder realizar una búsqueda dentro de la base de datos podemos utilizar el siguiente comando.

```python
#El siguiente comando rellena con un espacio para evitar cualquier tipo de problemas con datos no válidos
df=df.fillna('')
```

```python
#Buscara la palabra que asignes (Segundo corchete) dentro de la columna que le indiques (Primer corchete) y mostrará las coincidencias
#dentro de la base de datos.
#Por ejemplo, buscaremos todos los trabajadores que se llamen Juan dentro de la columna de Nombres
df[df['Nombre'].str.contains('Juan')]
```

```python
#Ahora haremos un ejemplo para cuando se tiene minusculas y mayusculas, igual buscará los becarios dentro de la columna de Trabajo
df[df['Trabajo'].str.contains('becario', case=False)]
```

```python
#Ahora haremos una mezcla de lo anterior,
#Se tienen minusculas y mayusculas, y buscará los becarios o ingenieros disponibles en la columna de Trabajo
df[df['Trabajo'].str.contains('becario|ingeniero', case=False, na=False, regex=True)]
```
