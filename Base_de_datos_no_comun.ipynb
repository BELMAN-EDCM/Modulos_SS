```python
df = pd.read_excel('/content/drive/MyDrive/Base de datos/TablaConstantesdeAntoine.xlsx', header=[20])
df.head()
#Si el encabezado fuesen filas combinadas, se tendria que indicar el número de filas en base a su número de excel
#Por ejemplo, si para este ejemplo, fuesen 2 filas combinadas, dentro del primer comando en lugar de ser "header=[20]"
#Sería "header=[20, 21]", mientras que el segundo seria df.head(2)
```

```python
#Ahora para los extremos.
#La función skipfooter nos permitirá indicar el número de filas que deseamos ignorar, esto, de abajo hacia arriba
df = pd.read_excel('/content/drive/MyDrive/Base de datos/TablaConstantesdeAntoine.xlsx', header=[20], skipfooter=1)
#Comporbamos imprimiendo las últimas 4 filas
df.tail(4)
```

```python
#Como dato extra, existe otra función que le permite a pandas identificar que número de
#columnas corresponden al indice, se llama "index_col" y la cual funciona de forma similar a header
#En nuestro ejemplo solo tenemos 1 columna que funciona como indice, por lo cual quedaría de la siguiente manera:
df = pd.read_excel('/content/drive/MyDrive/Base de datos/TablaConstantesdeAntoine.xlsx', header=[20], skipfooter=1, index_col=[0])
#Comporbamos imprimiendo las primeras 4 filas.
df.head(4)
```
